```python
import networkx as nx
```

```python
G=nx.DiGraph()
```

```python
G.add_edges_from([(1,2),(1,6),(2,3),(2,4),(2,6),(3,4),(3,5),(4,8),(4,9),(6,7)])
```

```python
nx.draw(G,with_labels=True)
```

```python
print("total number of nodes : " ,G.number_of_nodes())
```

```python
print("total number of nodes : " ,G.number_of_edges())
```

```python
print("list of nodes : " ,G.nodes())
```

```python
print("list of edges : " ,G.edges(data=True))
```

```python
print("degree for all nodes",dict(G.degree()))
```

```python
list(G.neighbors(1))
```

```python
nx.to_numpy_matrix(G)
```
